```python
import pandas as pd
import numpy as np
import seaborn as sea
import matplotlib.pyplot as plt

# Load your data
d1 = pd.read_csv('/content/auto-op-orders-last-3-months.csv')

# Count the occurrences of each 'current_status'
status_counts = d1['current_status'].value_counts()

# Create a bar plot for 'current_status' counts
plt.figure(figsize=(8, 6))

# Plot the total count
sea.barplot(x=status_counts.index, y=status_counts.values, palette='coolwarm', label='Total')

# Filter the DataFrame for 'returned' current_status
hist_1 = d1[d1['current_status'] == 'returned']

# Count the occurrences of 'returned' status
returned_counts = hist_1['current_status'].value_counts()

# Plot the count of 'returned' status
sea.barplot(x=returned_counts.index, y=returned_counts.values, palette='hot', label='Returned')

# Set labels and title
plt.xlabel('Current Status')
plt.ylabel('Count')
plt.title('Bar Graph of Returned Orders for Auto-OP')

# Show a legend
plt.legend()

# Display the plot
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sea
import matplotlib.pyplot as plt

# Load your data
d1 = pd.read_csv('/content/auto-op-orders-last-3-months.csv')

# Filter the DataFrame for 'returned' current_status
hist_1 = d1[d1['current_status'] == 'returned']

# Calculate the counts for 'returned' and 'not returned' current_status
returned_count = len(hist_1)
not_returned_count = len(d1) - returned_count

# Create a pie chart
labels = ['Returned', 'Not Returned']
sizes = [returned_count, not_returned_count]
colors = ['red', 'orange']
explode = (0.1, 0)  # Explode the 'Returned' slice for emphasis

plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Pie Chart Percentile of Returned Orders for Auto-OP')

# Display the pie chart
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
