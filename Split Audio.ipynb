```python
from pyAudioAnalysis import audioSegmentation
import os
from pydub import AudioSegment

def split_audio_by_speakers(audio_file_path):
    # Load the original audio file
    audio = AudioSegment.from_file(audio_file_path)

    # Duration of each chunk in milliseconds (30 seconds)
    chunk_duration = 30 * 1000

    # Create a folder to save the segmented audio files
    output_folder = 'speaker_segments'
    os.makedirs(output_folder, exist_ok=True)

    # Create dictionaries to store segments for each speaker
    speaker_segments = {0: [], 1: []}

    # Split audio into 30-second chunks and perform diarization on each chunk
    for i, start_time in enumerate(range(0, len(audio), chunk_duration)):
        chunk = audio[start_time:start_time + chunk_duration]

        # Export the chunk to a temporary WAV file
        temp_audio_file = f"temp_audio_chunk_{i}.wav"
        chunk.export(temp_audio_file, format="wav")

        try:
            # Perform speaker diarization on the chunk
            diarization_result = audioSegmentation.speaker_diarization(temp_audio_file, n_speakers=2)

            # Unpack the diarization result
            flagsInd, classesAll = diarization_result

            # Append the chunk to the respective speaker's segment list
            for idx, start_ind in enumerate(flagsInd):
                if idx < len(flagsInd) - 1:
                    start_sample = int(start_ind)
                    end_sample = int(flagsInd[idx + 1])

                    segment = audio[start_time + start_sample:start_time + end_sample]
                    speaker_segments[classesAll[idx]].append(segment)

        except Exception as e:
            print(f"Error processing chunk {i}: {e}")

        # Remove the temporary audio file
        os.remove(temp_audio_file)

    # Save combined segments for each speaker into a single output file
    for speaker, segments in speaker_segments.items():
        combined_segments = AudioSegment.empty()
        for segment in segments:
            combined_segments += segment

        output_file = os.path.join(output_folder, f'Speaker_{speaker}_Output.wav')
        combined_segments.export(output_file, format="wav")

    return f"Speaker diarization completed. Combined segments saved in '{output_folder}' folder as Speaker_{0, 1}_Output.wav"

# Usage of the function
audio_file_path = "/content/Voice_Data_1.wav"
result_message = split_audio_by_speakers(audio_file_path)
print(result_message)
```
